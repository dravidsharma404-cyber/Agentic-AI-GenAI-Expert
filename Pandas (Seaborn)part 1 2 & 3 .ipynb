## 🎯 Different Flier Types (Markers)

| Code  | Shape            |
| ----- | ---------------- |
| `"o"` | Circle (default) |
| `"D"` | Diamond 💎       |
| `"s"` | Square ⬛         |
| `"^"` | Triangle 🔺      |
| `"v"` | Triangle down 🔻 |
