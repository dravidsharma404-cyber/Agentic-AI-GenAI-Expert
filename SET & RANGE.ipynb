## # SUBSET ( ⊆ ) | SUPERSET ( ⊇ ) |  DISJOINT (No Common Elements)

```python
a = {1,2,3,4,5,6,7,8,9} 
b = {3,4,5,6,7,8} 
s7 = {10,20,30,40} 
```

```python
a.issubset(b)                # All b values are in "a" so a is superset Not Subset
```

```python
a.issuperset(b)                 # All b values are in "a" so a is superset
```

```python
a.isdisjoint(b)            # few values are in "a" & "b" are common so output is false
```

## 🔥 Full Comparison

| Concept  | Meaning      | Condition         |
| -------- | ------------ | ----------------- |
| Subset   | A inside B   | All A ∈ B         |
| Superset | B contains A | All A ∈ B         |
| Disjoint | No overlap   | No common element |
