# Complex Number in the cmath Module

```python
import cmath
```

```python
print(cmath.sqrt(2))
```

```python
print(cmath.sqrt(4))
```

```python
print(cmath.sin(2))
print(cmath.cos(2))
```

```python
z = 3 + 4j
print(cmath.phase(z))
```

```python
z = 3 + 4j
print(abs(z))
```

```python
z = 3 + 4j
print(cmath.polar(z))
```

```python
r = 5
theta = 0.927
print(cmath.rect(r, theta))
```

| Function       | Meaning              |
| -------------- | -------------------- |
| `sqrt()`       | Square root          |
| `sin(), cos()` | Trigonometry         |
| `phase()`      | Angle                |
| `polar()`      | (magnitude, angle)   |
| `rect()`       | Back to complex form |
