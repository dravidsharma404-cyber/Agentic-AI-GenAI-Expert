## 5. Loop (VERY IMPORTANT 🔥)

```python
d1={'name':'Dravid', 'Age':'30', 'City':'Agra'}
d2={'1':'One', '2':'Two', '3':'Three', '4':'Four', '5':'Five'}
```

```python
for k in d1.keys():
    print(k)
```

```python
for v in d1.values():
    print(v)
```

```python
for k,v in d1.items():
    print(k,v)
```

## 6. get() (MOST IMPORTANT 🔥🔥)

```python
d3=[1,1,1,2,2]

freq={}

for i in d3:
    freq[i]=freq.get(i,0)+1
    print(freq)
```

# If key not exist → create with value

| Function   | Use          |
| ---------- | ------------ |
| `d[key]`   | Get value    |
| `get()`    | Safe get     |
| `pop()`    | Remove       |
| `keys()`   | All keys     |
| `values()` | All values   |
| `items()`  | key + value  |
| `update()` | Add multiple |
| `clear()`  | Empty dict   |
