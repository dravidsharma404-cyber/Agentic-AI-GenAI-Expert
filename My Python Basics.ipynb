# Logical Operators (Combine Conditions)

| Operator | Meaning              |
| -------- | -------------------- |
| and      | Both conditions true | 1 and 1 = 1 rest all will 0
| or       | Any one true         | 0 or 0 = 0 rest call will 1
| not      | Reverse result       |


```python
| Condition 1 | Condition 2 | Result |
| ----------- | ----------- | ------ |
| True        | True        | True   |
| True        | False       | False  |
| False       | True        | False  |
| False       | False       | False  |
```

```python
a=2
b=4
```

```python
a>1 and b>3 # 2 greater than 1 (true) And 4 greater than 3 (true) = True
```

```python
a>1 and b<3 # 2 greater than 1 (true) And 4 less than 3 (false) = False
```

```python
a<1 and b>3 # 2 less than 1 (false) And 4 greater than 3 (true) = False
```

```python
a<1 and b<3 # 2 less than 1 (false) And 4 less than 3 (false) = False
```

# or Operator 
👉 Returns True if ANY ONE condition is True

```python
| Condition 1 | Condition 2 | Result |
| ----------- | ----------- | ------ |
| True        | True        | True   |
| True        | False       | True   |
| False       | True        | True   |
| False       | False       | False  |
```

```python
a=2
b=4
```

```python
a>1 or b>3 # 2 greater than 1 (true) or 4 greater than 3 (true) = True
```

```python
a>1 or b<3 # 2 greater than 1 (true) or 4 less than 3 (false) = True
```

```python
a<1 or b>3 # 2 less than 1 (false) or 4 greater than 3 (true) = True
```

```python
a<1 or b<3 # 2 less than 1 (false) or 4 less than 3 (false) = False
```

start → From where to begin (included)

end → Where to stop (NOT included ❌)

step → How many steps to jump (optional)

## Bitwise Operator

Topic: Bitwise Operators in Python
1️⃣ What is Bitwise Operator?

Bitwise operators work on binary numbers (0 and 1).

Example:

Decimal   Binary
5         101
6         110

Computers internally store numbers in binary, so bitwise operators manipulate these bits.

# Bitwise AND &

Rule

```python
1 & 1 → 1
1 & 0 → 0
0 & 1 → 0
0 & 0 → 0
```

```python
10&7
```

```python
10 = 1010
7  = 0111
---------
     0010
```
