```python
plt.plot(Salary[0], color='black', ls=':',marker='^',ms=7,label=Players[0])
plt.plot(Salary[1], color='purple', ls=':',marker='o',ms=5,label=Players[1])
plt.plot(Salary[2], color='red', ls=':',marker='<',ms=5,label=Players[2])
plt.xticks(list(range(0,10)),Sdict,rotation='horizontal')
plt.legend(loc='upper right',bbox_to_anchor=(1,0))
plt.rcParams['figure.figsize']=7,3
plt.show()
```

🎯 Why legend is going outside (right-bottom)
loc='upper center' → tells legend how to align
bbox_to_anchor=(1,0) → tells legend WHERE to go

👉 (1,0) means:

x = 1 → extreme right
y = 0 → bottom


✅ So final position becomes:
👉 Right + Bottom (outside the plot)

```python
plt.plot(Salary[0], color='black', ls=':',marker='^',ms=7,label=Players[0])
plt.plot(Salary[1], color='purple', ls=':',marker='o',ms=5,label=Players[1])
plt.plot(Salary[2], color='red', ls=':',marker='<',ms=5,label=Players[2])
plt.xticks(list(range(0,10)),Sdict,rotation='horizontal')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.rcParams['figure.figsize']=7,3
plt.show()
```
