# 1️⃣ List Functions

| Function    | How / Why                                   | Example                  |
| ----------- | ------------------------------------------- | ------------------------ |
| `append()`  | Add one element at end                      | `l=[1,2]; l.append(3)`   |
| `extend()`  | Add multiple elements from another iterable | `l=[1]; l.extend([2,3])` |
| `insert()`  | Insert element at specific index            | `l.insert(1,100)`        |
| `remove()`  | Remove first matching value                 | `l.remove(2)`            |
| `pop()`     | Remove element using index                  | `l.pop(1)`               |
| `clear()`   | Remove all elements                         | `l.clear()`              |
| `index()`   | Get index of value                          | `l.index(3)`             |
| `count()`   | Count occurrences                           | `l.count(2)`             |
| `sort()`    | Sort list ascending                         | `l.sort()`               |
| `reverse()` | Reverse list order                          | `l.reverse()`            |
| `copy()`    | Create shallow copy                         | `l2=l.copy()`            |


# 2️⃣ Tuple Functions

| Function  | How / Why              | Example                   |
| --------- | ---------------------- | ------------------------- |
| `count()` | Count occurrences      | `t=(1,2,2,3); t.count(2)` |
| `index()` | Find position of value | `t.index(3)`              |


# 3️⃣ Set Functions


| Function                 | How / Why                           | Example                       |
| ------------------------ | ----------------------------------- | ----------------------------- |
| `add()`                  | Add single element                  | `s.add(5)`                    |
| `update()`               | Add multiple elements               | `s.update([6,7])`             |
| `remove()`               | Remove element (error if not found) | `s.remove(3)`                 |
| `discard()`              | Remove element safely               | `s.discard(3)`                |
| `pop()`                  | Remove random element               | `s.pop()`                     |
| `clear()`                | Remove all elements                 | `s.clear()`                   |
| `union()`                | Combine two sets                    | `s1.union(s2)`                |
| `intersection()`         | Common elements                     | `s1.intersection(s2)`         |
| `difference()`           | Elements in first but not second    | `s1.difference(s2)`           |
| `symmetric_difference()` | Elements not common in both         | `s1.symmetric_difference(s2)` |
| `issubset()`             | Check if subset                     | `s1.issubset(s2)`             |
| `issuperset()`           | Check if superset                   | `s1.issuperset(s2)`           |


# 4️⃣ Dictionary Functions

| Function       | How / Why                   | Example                        |
| -------------- | --------------------------- | ------------------------------ |
| `keys()`       | Get all keys                | `d.keys()`                     |
| `values()`     | Get all values              | `d.values()`                   |
| `items()`      | Get key-value pairs         | `d.items()`                    |
| `get()`        | Safely access value         | `d.get('name')`                |
| `update()`     | Add or update values        | `d.update({'age':30})`         |
| `pop()`        | Remove specific key         | `d.pop('age')`                 |
| `popitem()`    | Remove last inserted item   | `d.popitem()`                  |
| `clear()`      | Remove all data             | `d.clear()`                    |
| `copy()`       | Copy dictionary             | `d2=d.copy()`                  |
| `setdefault()` | Add key if not present      | `d.setdefault('city','Delhi')` |
| `fromkeys()`   | Create dict with same value | `dict.fromkeys(['a','b'],0)`   |

# 1️⃣ Syntax Error

Happens when Python rules (syntax) are wrong

```python
if 5 > 2
    print("Hello")
```

Reason
: is missing after if.

```python
if 5 > 2:
    print("Hello")
```

Program runs without error but gives wrong output.

# ⚡ Common Runtime Errors in Python

| Error               | Reason                   | Example          |
| ------------------- | ------------------------ | ---------------- |
| `NameError`         | Variable not defined     | `print(x)`       |
| `TypeError`         | Wrong data type          | `5 + "A"`        |
| `ValueError`        | Invalid value            | `int("abc")`     |
| `IndexError`        | Index out of range       | `l[10]`          |
| `KeyError`          | Dictionary key not found | `d['age']`       |
| `ZeroDivisionError` | Division by zero         | `10/0`           |
| `AttributeError`    | Wrong method used        | `'abc'.append()` |
| `IndentationError`  | Wrong indentation        | wrong spacing    |
| `ImportError`       | Module not found         | `import xyz`     |
